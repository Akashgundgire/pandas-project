```python
import pandas as pd

dataset=pd.read_csv("netflix1.csv")
dataset
```

# First Cleaning the Null Values From The Dataset

```python
import numpy as np
newset=dataset.replace("Not Given",np.NaN)
newset
```

```python
cleaned=newset.dropna()
cleaned
```

# Question. Show countries in which highest and lowest films and TV shows were made? (films and tv shows combined)

```python
diff_content_per=cleaned.country.value_counts()
diff_content_per
```

# Question. Who are the top 5 directors with most movies and TV shows?

```python
directors_at_top=cleaned.groupby('director').type.count()
directors_at_top.sort_values(ascending=False)
```

# Question. How many movies were released after 1983?

```python
rndom1=cleaned.loc[(cleaned.type=='Movie') & (cleaned.release_year>1983)]
rndom1.type.count()
```

# Question. Show type of entertainment(movie or TV show) ,title and director of the film or TV show.

```python
#selecting only type,title and director column
cleaned.loc[:,['type','title','director']]
```

# Question. Show records where country is India and director is Suhas Kadav.

```python
cleaned.loc[(cleaned.country=='India') & (cleaned.director=='Suhas Kadav')]
```

# Question. Who is the director of the movie 'Sankofa'?

```python
cleaned.loc[cleaned.title=='Sankofa']
```

# Question. How many movies and tv shows were released on 1st October,2019?

```python
for_2019=cleaned.loc[cleaned.date_added=='11/1/2019']
for_2019
for_2019.type.count()
```

# Question. Show records which are unsuitable for children under 17.

```python
cleaned.loc[cleaned.rating=='TV-MA']

# Rating 'TV-MA'. This rating is given to those films which are not suitable for children under 17
```

```python
Question. Show movies which are from 'United States' and 'Germany'.
```

```python
cleaned.loc[(cleaned.type=='Movie') & (cleaned.country.isin(['United States','Germnay']))]
```

# Question. Shows movies and TV shows that has a date added between 1 January,2019 to 31 December,2019

```python
cleaned.date_added=cleaned.date_added.astype('datetime64')
cleaned

cleaned.loc[cleaned["date_added"].between("2019-01-01","2019-12-31")]
```
